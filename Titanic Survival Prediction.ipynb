```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

```python
titanic.drop(columns=['deck','alone','who','alive','class','embark_town','adult_male'],inplace=True)
```

```python
titanic.isnull().sum()
```

```python
titanic.iloc[:,3]=titanic.fillna(titanic.iloc[:,3].mean())
```

```python
titanic.iloc[:,7]=titanic.fillna(titanic.iloc[:,7].mode()[0])
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
titanic.iloc[:,7]=encoder.fit_transform(titanic.iloc[:,7])
```

```python
titanic.head(10)
```

```python
titanic.iloc[:,2]=encoder.fit_transform(titanic.iloc[:,2])
```

```python
titanic.head(10)
```

```python
x=titanic.drop(columns=['survived'])
y=titanic['survived']
```

```python
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
def models(x_train,y_train):
    from sklearn.linear_model import LogisticRegression
    log=LogisticRegression(random_state=0)
    log.fit(x_train,y_train)
    
    from sklearn.svm import SVC
    svm=SVC(kernel='linear',random_state=0)
    svm.fit(x_train,y_train)
    
    from sklearn.svm import SVC
    svm1=SVC(kernel='rbf',random_state=0)
    svm1.fit(x_train,y_train)
    
    
    from sklearn.tree import DecisionTreeClassifier
    tree=DecisionTreeClassifier(criterion='entropy',random_state=0)
    tree.fit(x_train,y_train)
    
    print('Logistic regression',log.score(x_train,y_train))
    print('svc linear regression',svm.score(x_train,y_train))
    print('tree regression',tree.score(x_train,y_train))
    print('svc rbf regression',svm1.score(x_train,y_train))
    
    return log,svm,svm1,tree
```

```python
model=models(x_train,y_train)
```
